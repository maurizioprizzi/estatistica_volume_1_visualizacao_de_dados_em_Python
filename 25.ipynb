```python
import pandas as pd
import matplotlib.pyplot as plt

# Importando os dados do .csv
df = pd.read_csv('populacao.csv')

# Definindo as faixas etárias como o índice do DataFrame para facilitar o plot
df.set_index('Faixa Etária', inplace=True)

# Criando o histograma
df['População'].plot(kind='bar', edgecolor='black')

plt.title('População por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('População')
plt.grid(axis='y')

plt.show()
```
