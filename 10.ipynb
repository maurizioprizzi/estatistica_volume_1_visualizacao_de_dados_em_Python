```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ler o arquivo .csv
df = pd.read_csv('depositos.csv')
```

```python
# Adicionar a terceira coluna com a soma progressiva acumulada
df['total_acumulado'] = df['deposito_anual'].cumsum()

# Ordenar os valores pela coluna 'idade' para criar o polígono de frequência acumulada
df = df.sort_values('idade')

# Criar o gráfico de polígonos de frequência acumulada
plt.plot(df['idade'], df['total_acumulado'], marker='o')

# Adicionar título e rótulos aos eixos
plt.title('Gráfico de Polígonos de Frequência Acumulada')
plt.xlabel('Idade')
plt.ylabel('Total Acumulado')

# Exibir o gráfico
plt.show()
```
