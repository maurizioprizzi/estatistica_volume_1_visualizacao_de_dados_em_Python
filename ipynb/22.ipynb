```python
import stemgraphic

# Idades de um grupo de pessoas
idades = [20, 24, 22, 24, 25, 26, 28, 22, 24, 23, 24, 25, 26, 27, 28, 29, 30, 32, 34, 35, 36, 37, 38, 39, 40, 42, 44, 45, 46, 47, 48, 49]

# Criando o diagrama de caule-e-folhas
fig, ax = stemgraphic.stem_graphic(idades)
```

```python
def stem_leaf_plot(data):
    # Organizando os dados
    data.sort()
    # Criando um dicionário para guardar os caules e as folhas
    stem_leaf = {}
    for number in data:
        stem, leaf = divmod(number, 10)
        # Adicionando as folhas aos caules correspondentes
        stem_leaf.setdefault(stem, []).append(leaf)
    # Imprimindo o gráfico
    for stem, leaves in stem_leaf.items():
        print(f"{stem} | {' '.join(map(str, leaves))}")

# Idades de um grupo de pessoas
ages = [20, 24, 22, 24, 25, 26, 28, 22, 24, 23, 24, 25, 26, 27, 28, 29, 30, 32, 34, 35, 36, 37, 38, 39, 40, 42, 44, 45, 46, 47, 48, 49]
stem_leaf_plot(ages)
```
