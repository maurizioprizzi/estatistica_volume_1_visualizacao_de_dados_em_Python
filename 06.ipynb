```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('jogos_invernais.csv')
```

```python
# df = dataframe
df
```

```python
# Agrupar as contagens por continente e país e somar o número de hospedagens
tabela_hospedagem = df.groupby(['Continente', 'País']).size().reset_index(name='Número de Hospedagens')
tabela_hospedagem = tabela_hospedagem.groupby('Continente')['Número de Hospedagens'].sum().reset_index()

# Ordenar a tabela em ordem decrescente pelo número total de hospedagens
tabela_hospedagem = tabela_hospedagem.sort_values(by='Número de Hospedagens', ascending=False)
```

```python
tabela_hospedagem
```

```python
# Exibir a tabela de hospedagem por continente sem o índice mesclado
print(tabela_hospedagem.to_string(index=False))
```

```python
# Criar o gráfico de barras
plt.bar(tabela_hospedagem['Continente'], tabela_hospedagem['Número de Hospedagens'])

# Adicionar título e rótulos aos eixos
plt.title('Número de Hospedagens dos Jogos Olímpicos Invernais por Continente')
plt.xlabel('Continente')
plt.ylabel('Número de Hospedagens')

# Exibir o gráfico
plt.show()
```
