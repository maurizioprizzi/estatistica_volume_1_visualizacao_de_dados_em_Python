```python
import pandas as pd
from IPython.display import display

# Carregar os dados do arquivo CSV
df = pd.read_csv('artista.csv')

# Calcular a soma dos ganhos por mês
df['Total'] = df.iloc[:, 1:].sum(axis=1)

# Adicionar uma linha 'Total' com a soma de cada coluna
df.loc['Total'] = df.sum(numeric_only=True)

# Formatar as células com duas casas decimais
df[['2019', '2020', '2021', '2022', 'Total']] = df[['2019', '2020', '2021', '2022', 'Total']].applymap('{:.2f}'.format)

# Substituir NaN na coluna "mês" por "Total"
df['mês'].fillna('Total', inplace=True)

# Exibir a tabela
display(df)
```
