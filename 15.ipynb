```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ler o arquivo CSV
df = pd.read_csv('artista.csv')

# Configurar as cores para cada ano
cores = ['blue', 'green', 'red', 'purple']

# Configurar a largura das barras
largura_barra = 0.2

# Obter os índices dos meses
indices = np.arange(len(df['mês']))

# Criar o gráfico de barras agrupadas por ano
for i, coluna in enumerate(df.columns[1:]):
    plt.bar(indices + (i * largura_barra), df[coluna], width=largura_barra, color=cores[i], label=coluna)

# Configurar os ticks do eixo x e rotacionar os rótulos dos meses
plt.xticks(indices + (largura_barra * 1.5), df['mês'], rotation='vertical')

# Adicionar título e rótulos aos eixos
plt.title('Ganhos Mensais por Ano')
plt.xlabel('Mês')
plt.ylabel('Ganhos (R$)')

# Adicionar legenda
plt.legend()

# Exibir o gráfico de barras agrupadas
plt.tight_layout()
plt.show()
```
