```python
# Vamos começar carregando o arquivo CSV com os dados da tabela.
# Certifique-se de ajustar o caminho do arquivo para o local correto.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar o arquivo CSV
dataframe = pd.read_csv('jogos.csv')
```

```python
# Agrupar as contagens por continente e país e somar o número de hospedagens
tabela_hospedagem = dataframe.groupby(['Continente', 'País']).size().reset_index(name='Número de Hospedagens')
tabela_hospedagem = tabela_hospedagem.groupby('Continente')['Número de Hospedagens'].sum().reset_index()

# Ordenar a tabela em ordem decrescente pelo número total de hospedagens
tabela_hospedagem = tabela_hospedagem.sort_values(by='Número de Hospedagens', ascending=False)
```

```python
# Exibir a tabela de hospedagem por continente sem o índice mesclado
print(tabela_hospedagem.to_string(index=False))
```

```python
# Criar o gráfico de pizza
plt.pie(tabela_hospedagem['Número de Hospedagens'], labels=tabela_hospedagem['Continente'], autopct='%1.1f%%')

# Adicionar título ao gráfico
plt.title('Porcentagem de Hospedagens dos Jogos Olímpicos por Continente')

# Exibir o gráfico
plt.show()
```
